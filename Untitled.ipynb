```python
# weather_openai.py
import json
import requests
from openai import OpenAI

client = OpenAI()  # picks up OPENAI_API_KEY from env

# 1. The actual Python function the LLM will be allowed to call
def get_weather(city: str) -> dict:
    # Step 1: turn city name into latitude/longitude
    geo = requests.get(
        "https://geocoding-api.open-meteo.com/v1/search",
        params={"name": city, "count": 1},
    ).json()
    if not geo.get("results"):
        return {"error": f"could not find city {city}"}
    lat = geo["results"][0]["latitude"]
    lon = geo["results"][0]["longitude"]

    # Step 2: get current weather for those coords
    wx = requests.get(
        "https://api.open-meteo.com/v1/forecast",
        params={"latitude": lat, "longitude": lon, "current_weather": True},
    ).json()
    return wx["current_weather"]

# 2. Describe the function to the LLM in OpenAI's format
openai_tools = [
    {
        "type": "function",
        "function": {
            "name": "get_weather",
            "description": "Get the current weather for a city",
            "parameters": {
                "type": "object",
                "properties": {
                    "city": {"type": "string", "description": "City name, e.g. Tokyo"},
                },
                "required": ["city"],
            },
        },
    }
]

# 3. The actual conversation loop
messages = [{"role": "user", "content": "What's the weather in Tokyo right now?"}]

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
    tools=openai_tools,
)

msg = response.choices[0].message

# 4. Did the LLM ask to call a tool?
if msg.tool_calls:
    messages.append(msg)  # remember what the LLM said
    for call in msg.tool_calls:
        args = json.loads(call.function.arguments)
        result = get_weather(**args)             # we run the Python function
        messages.append({
            "role": "tool",
            "tool_call_id": call.id,
            "content": json.dumps(result),
        })

    # 5. Send the result back so the LLM can answer the user
    final = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
    )
    print(final.choices[0].message.content)
else:
    print(msg.content)
```

```python
import json
import requests
from openai import OpenAI

client = OpenAI()


# 1. The actual Python function
def get_weather(city: str) -> dict:
    geo = requests.get(
        "https://geocoding-api.open-meteo.com/v1/search",
        params={"name": city, "count": 1},
    ).json()

    if not geo.get("results"):
        return {"error": f"could not find city {city}"}

    lat = geo["results"][0]["latitude"]
    lon = geo["results"][0]["longitude"]

    wx = requests.get(
        "https://api.open-meteo.com/v1/forecast",
        params={
            "latitude": lat,
            "longitude": lon,
            "current_weather": True,
        },
    ).json()

    return wx["current_weather"]


# 2. Tool definition
tools = [
    {
        "type": "function",
        "name": "get_weather",
        "description": "Get the current weather for a city",
        "parameters": {
            "type": "object",
            "properties": {
                "city": {
                    "type": "string",
                    "description": "City name, e.g. Tokyo",
                }
            },
            "required": ["city"],
        },
    }
]


# 3. Ask the model
response = client.responses.create(
    model="gpt-4.1-mini",
    input="What's the weather in Tokyo right now?",
    tools=tools,
)

# 4. Find function calls
for item in response.output:

    if item.type == "function_call":

        print(f"Model wants to call: {item.name}")

        args = json.loads(item.arguments)

        result = get_weather(**args)

        # 5. Send tool result back
        final_response = client.responses.create(
            model="gpt-4.1-mini",
            previous_response_id=response.id,
            input=[
                {
                    "type": "function_call_output",
                    "call_id": item.call_id,
                    "output": json.dumps(result),
                }
            ],
        )

        print(final_response.output_text)
```

```python
import os

def list_downloads():
    return os.listdir(".")
```

```python
tools = [
    {
        "type": "function",
        "name": "list_downloads",
        "description": "List files in the Downloads folder",
        "parameters": {
            "type": "object",
            "properties": {}
        }
    }
]
```

```python
response = client.responses.create(
    model="gpt-4.1-mini",
    input="What files are present in my Downloads folder?",
    tools=tools
)

print(response.output)
```

```python
tool_call = response.output[0]

result = list_downloads()
```
